```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')
#mpl.use('pdf')
mpl.rcParams['xtick.major.size'] = 8
mpl.rcParams['xtick.major.width'] = 2
mpl.rcParams['ytick.major.size'] = 8
mpl.rcParams['ytick.major.width'] = 2
mpl.rcParams['axes.linewidth']= 2
file=['sasa']#,'hbonds_pnw','hbonds_pnp','radius_of_gyration','rmsd','sasa']
for a in file:
    data1=np.genfromtxt('../k12l12_silica/simulation_results/'+a+'_3.dat')
    data2=np.genfromtxt('../lysine_silica/simulation_results/'+a+'_3.dat')
    data3=np.genfromtxt('../silica/simulation_results/'+a+'_3.dat')
    plt.figure(facecolor='white', figsize=(10,7))
    plt.title(a,weight='bold')
    numbers1=np.arange(0,(max(data1[:,0])+1)/10,0.1)
    numbers2=np.arange(0,(max(data2[:,0])+1)/10,0.1)
    numbers3=np.arange(0,(max(data3[:,0])+1)/10,0.1)
    plt.plot(numbers1, data1[:,1], color='green', label=r"$K_{12}L_{12}-Silica$")
    plt.plot(numbers2, data2[:,1], color='blue', label=r"$K_{20}-Silica$")
    plt.plot(numbers3, data3[:,1], color='red', label=r"$Silica$")
    plt.xticks(np.arange(0,325,25),rotation=0,fontsize=18,weight='bold')
    plt.xlim(0,300)
    plt.yticks(fontsize=18,weight='bold')
    plt.xlabel("Time (ns)",weight='bold',fontsize=18)
    plt.ylabel("SASA (Å^2)",weight='bold',fontsize=18) #"RGYR(Å)"
    plt.legend(numpoints = 1,prop=dict(weight='bold',size=18),loc='best')
    # Set the minor ticks
    ax = plt.gca()
    x_minor_locator = plt.MultipleLocator(5)
    #y_minor_locator = plt.MultipleLocator(0.5)
    ax.xaxis.set_minor_locator(x_minor_locator)
    #ax.yaxis.set_minor_locator(y_minor_locator)
    ax.tick_params(which='both', width=2)
    ax.tick_params(which='major', length=10)
    ax.tick_params(which='minor', length=5, color='black')
    plotfile=a+'_3.png'
    plt.savefig(plotfile, dpi=300, bbox_inches='tight')
```
